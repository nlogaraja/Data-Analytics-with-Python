```python
l_df = pd.DataFrame({'Alphabet' : ['a', 'b', 'c', 'd'],
                     'Number' : [1, 2, 6, 5,],
                     'AlphaNumeric' : ['a1', 'b2', 'c3', 'd4']})

r_df = pd.DataFrame({'Alphabet' : ['a', 'b', 'c', 'd'],
                     'Number' : [5, 6, 7, 8],
                     'AlphaNumeric' : ['e5', 'f6', 'g7', 'h8']})
```

```python
merged_df = pd.merge(l_df, r_df, on = 'Number', how = 'right')
merged_df
```

```python
df1 = pd.DataFrame({'keyleft' : [0, 1, 2, 3],
                    'Aleft' : ['A0', 'A1', 'A2', 'A3']})

df2 = pd.DataFrame({'keyright' : [0, 1],
                    'Aright' : ['A0', 'A1']})

joined_df = df1.join(df2)
joined_df
```
