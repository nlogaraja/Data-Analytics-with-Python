```python
#import necessory libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset (built-in dataset in Seaborn)
iris = sns.load_dataset('iris')
```

```python
# Display the first few rows of the dataset
print("Sample of the Iris dataset:")
print(iris.head())
```

```python
#check for null values
iris.isnull().sum()
```

```python
# Summary statistics of numerical features
print("\nSummary Statistics:")
print(iris.describe())
```

```python
# Pair plot for visualizing relationships between features
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

```python
# Box plot for individual features
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=iris)
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=iris)
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=iris)
plt.suptitle("Box Plots of Iris Dataset Features")
plt.show()
```

```python
# Correlation matrix
correlation_matrix = iris.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
